### Q3: Count of Movies & TV Shows (Bar Graph)

```python
# Count the number of Movies and TV Shows
category_counts = data['Category'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Number of Movies & TV Shows')
plt.xlabel('Category')
plt.ylabel('Count')

# Add value labels on top of each bar
for i, v in enumerate(category_counts):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Print the counts
print(category_counts)
```

### Q7: Movies (Comedies) or Shows from UK

```python
# filter the data based on the conditions
comedies_movies = data[(data['Category'] == 'Movie') & (data['Type'].str.contains('Comedies'))]
uk_movies_shows = data[data['Country'] == 'United Kingdom']

# visualization dataframe
visualization_data = pd.DataFrame({
    'Condition': ['Comedies Movies', 'Movies/Shows from UK'],
    'Count': [len(comedies_movies), len(uk_movies_shows)]
})

# style for the plot
sns.set(style="whitegrid")

# Plot the data using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Condition', y='Count', data=visualization_data, palette='rocket')

# Set plot title and labels
plt.title('Comparison of Comedies Movies and Movies/Shows from UK', fontsize=14)
plt.xlabel('Condition', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

## Q9.2: 'R' rated TV Shows after 2018

```python
# convert Release_Date to datetime if not already done
data['Release_Date'] = pd.to_datetime(data['Release_Date'], errors='coerce')

# extract year into a new column
data['Release_Year'] = data['Release_Date'].dt.year

# filter for TV Shows with 'R' rating after 2018
tv_shows_r = data[(data['Rating'] == 'R') & (data['Release_Year'] > 2018) & (data['Category'] == 'TV Show')]

# count the result
count_tv_shows_r = tv_shows_r.shape[0]

print(f"Number of TV Shows with 'R' rating after 2018: {count_tv_shows_r}")
```

### Q10: Maximum Duration

```python
# convert duration column for movies to integers
data['Duration_Minutes'] = data['Duration'].str.extract('(\d+)').astype(float)

# filter out movies only and find the maximum duration
max_duration_movie = data[data['Category'] == 'Movie']['Duration_Minutes'].max()

print(f"The maximum duration of a movie on Netflix is: {max_duration_movie} minutes")
```

```python
# Maximum number of seasons
# Extract number of seasons for TV Shows
data['Seasons'] = data['Duration'].apply(lambda x: x.split(' ')[0] if 'Season' in x else None)

# Filter out TV Shows and find the maximum number of seasons
max_seasons_tv_show = data[data['Category'] == 'TV Show']['Seasons'].astype(float).max()

print(f"The TV Show with the most seasons on Netflix has: {max_seasons_tv_show} seasons")
```

### Q13: Movies (Dramas) or TV Shows (Kids' TV)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the data
dramas_movies = data[(data['Category'] == 'Movie') & (data['Type'].str.contains('Dramas'))]
kids_tv = data[(data['Category'] == 'TV Show') & (data['Type'].str.contains('Kids'))]

# dataFrame for visualization
visualization_data = pd.DataFrame({
    'Category': ['Dramas Movies', 'Kids TV Shows'],
    'Count': [len(dramas_movies), len(kids_tv)]
})

# Set the style for the plot
sns.set(style="whitegrid")

# Plot the data using seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Count', data=visualization_data, palette='viridis')

# Set plot title and labels
plt.title('Comparison of Dramas Movies and Kids TV Shows on Netflix', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```
