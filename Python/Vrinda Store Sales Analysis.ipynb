## Data Cleaning

###  Standadizing Column Values

```python
# Standardize Gender column
vrinda_df["Gender"] = vrinda_df["Gender"].str.strip().str.lower()  # Remove spaces & make lowercase
vrinda_df["Gender"] = vrinda_df["Gender"].replace({
    "male": "Men",
    "m": "Men",
    "men": "Men",
    "female": "Women",
    "f": "Women",
    "women": "Women",
    "w": "Women"
})

print(vrinda_df["Gender"].unique())

# Quantity column 
word_to_num = {
    "one": 1,
    "two": 2,
}

vrinda_df["Qty"] = vrinda_df["Qty"].astype(str).str.lower().map(word_to_num).fillna(vrinda_df["Qty"])

vrinda_df["Qty"] = pd.to_numeric(vrinda_df["Qty"], errors="coerce")

print(vrinda_df["Qty"].unique())

# Save cleaned data
vrinda_df.to_csv("vrinda_cleaned.csv", index=False)

print("Gender column standardized successfully!")
print("Quantity column standardized successfully!")
```

### Categorical vs. Numerical Correlation (Point Biserial)

```python
from scipy.stats import pointbiserialr

corr, p_value = pointbiserialr(vrinda_df["gender"].map({"Men": 1, "Women": 0}), vrinda_df["amount"])
print(f"Point-Biserial Correlation: {corr}, P-value: {p_value}")

```

### Customer Retention Analysis

```python
# first and last purchase for each customer
vrinda_df["date"] = pd.to_datetime(vrinda_df["date"])

customer_activity = vrinda_df.groupby("cust id").agg(
    First_Purchase=("date", "min"),
    Last_Purchase=("date", "max"),
    Total_Orders=("cust id", "count")
).reset_index()

print(customer_activity.head())
```

```python
# calculating retention rate
vrinda_df["date"] = pd.to_datetime(vrinda_df["date"])

customer_activity = vrinda_df.groupby("cust id").agg(
    First_Purchase=("date", "min"),
    Last_Purchase=("date", "max"),
    Total_Orders=("cust id", "count")
).reset_index()
```

```python
# Extract year-month from first purchase date
customer_activity["First_Purchase_Month"] = customer_activity["First_Purchase"].dt.to_period("M")

# Count customers per month
monthly_customers = customer_activity.groupby("First_Purchase_Month")["cust id"].nunique().reset_index()
monthly_customers.columns = ["Month", "New_Customers"]

# Shift column to track returning customers
monthly_customers["Returning_Customers"] = monthly_customers["New_Customers"].shift(1).fillna(0)

# Calculate retention rate per month
monthly_customers["Retention_Rate"] = (monthly_customers["Returning_Customers"] / monthly_customers["New_Customers"].shift(1)) * 100

# Fill NaN values with 0
monthly_customers.fillna(0, inplace=True)

print(monthly_customers.head())
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_customers["Month"].astype(str), y=monthly_customers["Retention_Rate"], marker="o", color="red")

plt.title(" Monthly Customer Retention Rate", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Retention Rate (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### The data reveals a remarkable and sustained improvement in customer retention at Vrinda Store from January 2022 to December 2022. After starting at 0% in January 2022, the retention rate quickly escalates to 100% by February 2022 and remains consistently high through the entire year. This significant increase and sustained retention rate highlight the effectiveness of the store's customer retention strategies, underscoring their success in building and maintaining strong customer loyalty over time.
